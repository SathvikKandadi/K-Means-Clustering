```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import  KMeans
```

```python
import pandas as pd
df = pd.read_excel("/AMCAT-random.xlsx")
df = df.head(100)
df
```

```python
plt.scatter(df['Automata'],df['Automata Fix'],)
plt.xlabel("Automata fix")
plt.ylabel("Automata")
plt.plot()
```

```python
df.columns
```

```python
k = KMeans(n_clusters = 3)
y = k.fit_predict(df[['Automata','Automata Fix']])
y
```

```python
df["clusters"] = y
df.head()
```

```python
df1 = df[df["clusters"] == 0]
df2 = df[df["clusters"] == 1]
df3 = df[df["clusters"] == 2]
plt.scatter(df1['Automata'],df1['Automata Fix'],color = "red")
plt.scatter(df2['Automata'],df2['Automata Fix'], color = "blue")
plt.scatter(df3['Automata'],df3['Automata Fix'],color  = "green")
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],marker = '*',color = "black")
```

```python
sse = []
kr = range(1,10)
for i in kr:
  k = KMeans(n_clusters = i)
  k.fit(df[['Automata','Automata Fix']])
  sse.append(k.inertia_)
```

```python

for i in range(1,len(sse)):
  print("k=",i,"sse",sse[i])
```

```python
plt.plot(kr,sse)
plt.xlabel("k")
plt.ylabel("sum of square errors")
```
